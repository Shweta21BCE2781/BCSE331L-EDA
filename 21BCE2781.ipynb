# Outlier detection and filtering

We can check for outliers using boxplots

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure()
sns.boxplot(data=df[['Age', 'DailyRate','DistanceFromHome']])
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df[['Education','EmployeeCount','EmployeeNumber', 'MonthlyIncome', 'YearsAtCompany']])
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df[['EnvironmentSatisfaction','HourlyRate','JobInvolvement', 'JobLevel', 'JobSatisfaction']])
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df[['MonthlyIncome','MonthlyRate','NumCompaniesWorked', 'PercentSalaryHike', 'PerformanceRating']])
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df[['RelationshipSatisfaction','StandardHours','StockOptionLevel', 'TotalWorkingYears', 'TrainingTimesLastYear']])
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df[['WorkLifeBalance','YearsAtCompany','YearsInCurrentRole', 'YearsSinceLastPromotion', 'YearsWithCurrManager']])
plt.show()
```

There are outliers in 'YearsAtCompany','YearsInCurrentRole', 'YearsSinceLastPromotion', 'YearsWithCurrManager','StockOptionLevel', 'TotalWorkingYears', 'TrainingTimesLastYear','MonthlyIncome','NumCompaniesWorked', 'PerformanceRating','MonthlyIncome'


The outliers are outside the range of:
( Q1-1.5IQR , Q3+1.5IQR )

So we can remove them using this formula

```python
columns = ['YearsAtCompany','YearsInCurrentRole', 'YearsSinceLastPromotion', 'YearsWithCurrManager','StockOptionLevel', 'TotalWorkingYears', 'TrainingTimesLastYear','MonthlyIncome','NumCompaniesWorked', 'PerformanceRating','MonthlyIncome']
for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1  # Interquartile range

        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR

        df = df[(df[col] >= lower) & (df[col] <= upper)]
```
