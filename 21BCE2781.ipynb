There are outliers in 'YearsAtCompany','YearsInCurrentRole', 'YearsSinceLastPromotion', 'YearsWithCurrManager','StockOptionLevel', 'TotalWorkingYears', 'TrainingTimesLastYear','MonthlyIncome','NumCompaniesWorked', 'PerformanceRating','MonthlyIncome'


The outliers are outside the range of:
( Q1-1.5IQR , Q3+1.5IQR )

So we can remove them using this formula

```python
columns = ['YearsAtCompany','YearsInCurrentRole', 'YearsSinceLastPromotion', 'YearsWithCurrManager','StockOptionLevel', 'TotalWorkingYears', 'TrainingTimesLastYear','MonthlyIncome','NumCompaniesWorked', 'PerformanceRating','MonthlyIncome']
for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1  # Interquartile range

        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR

        df = df[(df[col] >= lower) & (df[col] <= upper)]
```

# Module 3

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.display import display, Markdown

# Select only numerical columns
numerical_df = df.select_dtypes(include=['number'])
```

# Central Tendency Measures
###  Univariate Analysis


```python
central_tendency = numerical_df.agg(['mean', 'median'])
display(central_tendency)
```

### Dispersion Measures

```python
dispersion = numerical_df.agg(['min', 'max', 'var', 'std'])
display(dispersion)
```

### Interquartile Range (IQR)

```python
iqr = numerical_df.quantile([0.25, 0.75])
iqr.loc['IQR'] = iqr.loc[0.75] - iqr.loc[0.25]
display(iqr)
```

### Histograms of Numerical Variables

```python
num_cols = len(numerical_df.columns)
num_rows = (num_cols + 3) // 4

# subplots
fig, axes = plt.subplots(nrows=num_rows, ncols=4, figsize=(16, 4 * num_rows))
axes = axes.flatten()

for i, col in enumerate(numerical_df.columns):
    sns.histplot(numerical_df[col], bins=30, kde=True, ax=axes[i])
    axes[i].set_title(f"Histogram of {col}")
    axes[i].set_xlabel("")
    axes[i].set_ylabel("Frequency")

for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

# Bivariate Analysis
### Correlation Matrix

```python
correlation_matrix = numerical_df.corr()
display(correlation_matrix)
```


**Correlation Levels :**


1. No Correlation: Variables that have a
correlation close to zero.

2. Weak Correlation: Variables with a correlation between 0.1 and 0.3.

3. Strong Correlation: Variables with a correlation above 0.7 (or below -0.7).

### Correlation Heatmap

```python
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("21BCE2781 - Correlation Heatmap")
plt.show()
```
